```python
import os
import tensorflow as tf
import matplotlib.pyplot as plt

# 1. Definir os caminhos corretos com base na sua estrutura no Drive
base_path = '/content/drive/MyDrive/projeto_segmentacao/Dataset/Train'
image_dir = os.path.join(base_path, 'Image')
mask_dir = os.path.join(base_path, 'Mask')

# 2. Listar os arquivos
images = sorted([os.path.join(image_dir, f) for f in os.listdir(image_dir)])
masks = sorted([os.path.join(mask_dir, f) for f in os.listdir(mask_dir)])


# 3. Função pura em TensorFlow para ler e processar os pixels
def load_and_preprocess_image(image_path, mask_path):
    # --- Processando a Imagem Original ---
    img_raw = tf.io.read_file(image_path)
    img = tf.image.decode_jpeg(img_raw, channels=3) 
    img = tf.image.resize(img, [256, 256])
    img = tf.image.convert_image_dtype(img, tf.float32)

    # --- Processando a Máscara (Gabarito) ---
    mask_raw = tf.io.read_file(mask_path)
    mask = tf.image.decode_png(mask_raw, channels=1)
    mask = tf.image.resize(mask, [256, 256], method='nearest') # 'nearest' evita criar pixels intermediários
    mask = tf.image.convert_image_dtype(mask, tf.float32)

    return img, mask

img_tensor, mask_tensor = load_and_preprocess_image(images[0], masks[0])
print("--- REPRESENTAÇÃO DO TENSORFLOW ---")
print(f"Shape da imagem (pixels): {img_tensor.shape} | Tipo: {img_tensor.dtype}")
print(f"Shape da máscara (pixels): {mask_tensor.shape} | Tipo: {mask_tensor.dtype}")
```
