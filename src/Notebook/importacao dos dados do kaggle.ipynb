```python
import os
import json
!pip install python-dotenv
from dotenv import load_dotenv

# 1. Carregar as variáveis do arquivo .env
# o .env deve esta dentro do drive na pasta do colab para ele achar onde este
load_dotenv()

# Recuperando os valores salvos no seu .env
kaggle_username = os.getenv('KAGGLE_USERNAME')
kaggle_api_token = os.getenv('KAGGLE_API_TOKEN')

# Validação simples para garantir que o Python leu o arquivo .env com sucesso
if not kaggle_username or not kaggle_api_token:
    raise ValueError("Erro: Não foi possível ler 'KAGGLE_USERNAME' ou 'KAGGLE_API_TOKEN' do arquivo .env. Verifique se o arquivo está na raiz do projeto.")

# Configurando as variáveis no ambiente do Linux para a API do Kaggle
os.environ['KAGGLE_USERNAME'] = kaggle_username
os.environ['KAGGLE_API_TOKEN'] = kaggle_api_token

# 2. Garantindo que o cliente clássico do Kaggle também encontre o token no arquivo físico correto
kaggle_path = os.path.expanduser('~/.kaggle')
os.makedirs(kaggle_path, exist_ok=True)

# Salvando no novo formato de acesso da API
with open(os.path.join(kaggle_path, 'access_token'), 'w') as f:
    f.write(kaggle_api_token)

# Mantendo compatibilidade com o formato clássico JSON
token = {"username": kaggle_username, "key": kaggle_api_token}
with open(os.path.join(kaggle_path, 'kaggle.json'), 'w') as f:
    json.dump(token, f)

# Permissões restritas exigidas pelo Kaggle (apenas o seu usuário lê/escreve)
os.chmod(os.path.join(kaggle_path, 'kaggle.json'), 0o600)
os.chmod(os.path.join(kaggle_path, 'access_token'), 0o600)

# 3. Instalar/Atualizar a biblioteca do Kaggle
!pip install -q --upgrade kaggle

# 4. Criar a pasta de destino
output_dir = './dataset_segmentation'
os.makedirs(output_dir, exist_ok=True)

# 5. Baixar o dataset da competição
!kaggle competitions download -c a0-2025-medical-image-segmentation -p {output_dir}

print("\n--- Conteúdo da pasta final ---")
print(os.listdir(output_dir))
```

```python
import zipfile
import os
import shutil

# 1. Definir caminhos no seu Google Drive para salvar permanentemente
drive_dataset_dir = '/content/drive/MyDrive/projeto_segmentacao'
os.makedirs(drive_dataset_dir, exist_ok=True)

# 2. Mover o arquivo zip para o Google Drive
zip_original = './dataset_segmentation/a0-2025-medical-image-segmentation.zip'
zip_destino = os.path.join(drive_dataset_dir, 'dataset.zip')

print("Movendo arquivo para o Google Drive...")
shutil.move(zip_original, zip_destino)

# 3. Descompactar direto no Drive
print("Descompactando os arquivos no seu Google Drive (isso pode levar um minutinho)...")
with zipfile.ZipFile(zip_destino, 'r') as zip_ref:
    zip_ref.extractall(drive_dataset_dir)

# 4. Remover o zip para não ocupar espaço extra no Drive
os.remove(zip_destino)

print(f"Sucesso! Dados extraídos permanentemente em: {drive_dataset_dir}")
print("Conteúdo da pasta:")
print(os.listdir(drive_dataset_dir))
```
